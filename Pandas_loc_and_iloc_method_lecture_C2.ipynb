```python
import pandas as pd
#read csv file
data=pd.read_csv("/content/drive/MyDrive/pandas database/alcohol-consumption.csv",index_col="country")
#display csv file
print(data)
```

```python
#get first 3 rows by iloc (index location) method
data.iloc[0:4]
```

```python
#Multiple parameters
# retrieving rows by loc method
rows = data.loc[["Estonia", "Lithuania"]]
  
# checking data type of rows
print(type(rows))
  
# display
rows
```
